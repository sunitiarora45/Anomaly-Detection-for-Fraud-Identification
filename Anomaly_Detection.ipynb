```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split

# Step 1: Loading the dataset
creditcard_fraud_df = pd.read_csv('creditcard.csv')

# Display the first few rows to check the data
creditcard_fraud_df.head()
```

```python
# Checking the size of the dataset
print("Credit Card Fraud Analysis Dataset has {} samples with {} features each.".format(*creditcard_fraud_df.shape))

# Data Exploration
creditcard_fraud_df.describe()
```

```python
# Step 2: Data preprocessing

# Check for missing values
print(creditcard_fraud_df.isnull().sum())

# Normalize the 'Time' and 'Amount' features
scaler = StandardScaler()
creditcard_fraud_df['Time'] = scaler.fit_transform(creditcard_fraud_df['Time'].values.reshape(-1, 1))
creditcard_fraud_df['Amount'] = scaler.fit_transform(creditcard_fraud_df['Amount'].values.reshape(-1, 1))
```

```python
# Explore the dataset to understand the distribution of normal and fraudulent transactions

# Count the number of normal (Class=0) and fraudulent (Class=1) transactions
normal_count = creditcard_fraud_df[creditcard_fraud_df['Class'] == 0]['Class'].count()
fraud_count = creditcard_fraud_df[creditcard_fraud_df['Class'] == 1]['Class'].count()

# Plot the distribution
plt.figure(figsize=(5, 5))
plt.bar(['Normal', 'Fraud'], [normal_count, fraud_count])
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
plt.title('Normal vs. Fraud Transactions')
plt.show()

# Calculate the percentage of fraudulent transactions in the dataset
fraud_percentage = (fraud_count / (normal_count + fraud_count)) * 100
print(f"Fraudulent transactions account for approximately {fraud_percentage:.3f}% of all transactions.")
```

```python
def plot_roc_curve(fpr, tpr, roc_auc, algorithm_name):
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) - ' + algorithm_name)
    plt.legend(loc='lower right')
    plt.show()

def isolation_forest_evaluation(X, y):
    # Count the number of normal (Class=0) and fraudulent (Class=1) transactions
    normal_count = y[y == 0].count()
    fraud_count = y[y == 1].count()

    # Calculate the contamination factor for Isolation Forest
    outlier_fraction = fraud_count / (normal_count + fraud_count)

    # Instantiate and fit the Isolation Forest model
    isolation_forest_model = IsolationForest(contamination=outlier_fraction)
    isolation_forest_model.fit(X)

    # Predict outliers using the trained Isolation Forest model
    isolation_outliers_pred = isolation_forest_model.predict(X)
    isolation_outliers = (isolation_outliers_pred == -1)  # -1 represents outlier predictions

    # Calculate the classification report
    class_names = ['Normal', 'Fraud']
    report = classification_report(y, isolation_outliers, target_names=class_names)

    # Calculate the ROC AUC score for Isolation Forest
    roc_auc = roc_auc_score(y, isolation_outliers)

    # Calculate the ROC curve for Isolation Forest
    fpr, tpr, _ = roc_curve(y, isolation_outliers)

    # print("Isolation Forest: {}".format(isolation_outliers.sum()))
    print("Isolation Forest Classification Report: \n")
    print("Accuracy Score:", accuracy_score(y, isolation_outliers))
    print("ROC AUC Score:", roc_auc, "\n")
    print(report)

    # Plot the ROC curve
    plot_roc_curve(fpr, tpr, roc_auc, "Isolation Forest")

def local_outlier_factor_evaluation(X, y):
    
    # Count the number of normal (Class=0) and fraudulent (Class=1) transactions
    normal_count = y[y == 0].count()
    fraud_count = y[y == 1].count()

    # Calculate the contamination factor for Local Outlier Factor
    outlier_fraction = fraud_count / (normal_count + fraud_count)
    
    # Instantiate and fit the Local Outlier Factor model on the entire dataset (X)
    lof_model = LocalOutlierFactor(n_neighbors=150, algorithm='auto', leaf_size=30, metric='minkowski',
                                   p=2, metric_params=None, contamination=outlier_fraction)

    # Predict the labels (1 for anomalies, -1 for normal instances)
    lof_outliers_pred = lof_model.fit_predict(X)

    # Convert the LOF predictions to binary (0 for normal, 1 for fraudulent) as LOF outputs -1 for anomalies
    lof_outliers = (lof_outliers_pred == -1).astype(int)

    # Calculate the classification report
    class_names = ['Normal', 'Fraud']
    report = classification_report(y, lof_outliers, target_names=class_names)

    # Calculate the ROC AUC score for Local Outlier Factor
    roc_auc = roc_auc_score(y, lof_outliers)

    # Calculate the ROC curve for Local Outlier Factor
    fpr, tpr, _ = roc_curve(y, lof_outliers)

    # print("Local Outlier Factor: {}".format(lof_outliers.sum()))
    print("Local Outlier Factor Classification Report: \n")
    print("Accuracy Score:",accuracy_score(y, lof_outliers))
    print("ROC AUC Score:", roc_auc, "\n")
    print(report)

    # Plot the ROC curve
    plot_roc_curve(fpr, tpr, roc_auc, "Local Outlier Factor")

# Sample a smaller subset of the data for evaluation
creditcard_fraud_sampled = creditcard_fraud_df.sample(frac=0.1, random_state=1)

# Separate features and target variable
X = creditcard_fraud_sampled.drop('Class', axis=1)
y = creditcard_fraud_sampled['Class']

# Call the functions to evaluate the two algorithms separately
isolation_forest_evaluation(X, y)
local_outlier_factor_evaluation(X, y)
```
